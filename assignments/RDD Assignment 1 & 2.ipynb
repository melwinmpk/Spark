```python
from pyspark import SparkConf, SparkContext
```

```python
Sparkconf = SparkConf().setAppName("rddlearning").setMaster("local")
sc = SparkContext(conf=Sparkconf)
```

## Assignmet 1
<pre>
1) Find count of CLOSED, PENDING_PAYMENT, ON_HOLD & COMPLETE orders?
Filename: orders.txt
</pre>

```python
file = sc.textFile("file:///home/saif/LFS/datasets/orders.txt")
```

```python
file.collect()
```

```python
header = file.first()
```

```python
filtered_data = file.filter(lambda x: x!=header and x.split(",")[3] in ['CLOSED','PENDING_PAYMENT','ON_HOLD','COMPLETE'])
```

```python
filtered_data.take(5)
```

```python
mapped_data = filtered_data.map(lambda x: (x.split(",")[3],1))
```

```python
mapped_data.take(5)
```

```python
counting = mapped_data.reduceByKey(lambda a,b:a+b)
```

```python
counting.collect()
```

## Assignment 2

<pre>
1) From the movies file display 1st & 2nd column & split 2nd column value by ('|') and repeat on every row.

Output:-
(Toy Story (1995),Adventure)
(Toy Story (1995),Animation)
(Toy Story (1995),Children)
(Toy Story (1995),Comedy)
(Toy Story (1995),Fantasy)
</pre>

```python
moviesrdd = sc.textFile("file:///home/saif/LFS/datasets/movies.csv")
moviesrdd.take(5)
```

```python
moviesrdd_header = moviesrdd.first()
```

```python
moviesrdd_without_header = moviesrdd.filter(lambda x: x!= moviesrdd_header)
moviesrdd_without_header.take(5)
```

```python
rdd_mapped_data = moviesrdd_without_header.map(lambda x: [(x.split(",")[1],i) for i in (x.split(",")[2]).split("|")])
rdd_mapped_data.take(5)
```

```python
for rows in rdd_mapped_data.collect():
    for row in rows:
        print(row)
```
